```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/smart_city_energy.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Zone analysis
zone_avg = df.groupby('zone')['energy_consumption_kwh'].mean().sort_values(ascending=False)
zone_avg.plot(kind='bar', color='skyblue', title='Average Energy Consumption by Zone')
plt.ylabel('Average kWh')
plt.show()

# Monthly trend
df['month'] = df['timestamp'].dt.month
monthly = df.groupby('month')['energy_consumption_kwh'].mean()
monthly.plot(kind='line', marker='o', color='orange', title='Monthly Energy Usage Trend')
plt.ylabel('kWh')
plt.show()

# Hourly pattern
df['hour'] = df['timestamp'].dt.hour
hourly = df.groupby('hour')['energy_consumption_kwh'].mean()
hourly.plot(kind='line', marker='o', color='green', title='Hourly Energy Consumption Pattern')
plt.ylabel('kWh')
plt.show()
```
