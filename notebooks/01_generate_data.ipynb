```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define zones and hourly timestamps
zones = ['North', 'South', 'East', 'West', 'Central']
hours = pd.date_range('2024-01-01', '2024-12-31 23:00:00', freq='H')

# Create base data
data = {
    'timestamp': np.repeat(hours, len(zones)),
    'zone': np.tile(zones, len(hours)),
    'energy_consumption_kwh': np.random.uniform(100, 500, len(hours) * len(zones))
}

df = pd.DataFrame(data)

# Add seasonal effect â€” higher consumption in summer months
df['month'] = df['timestamp'].dt.month
df.loc[df['month'].isin([4, 5, 6]), 'energy_consumption_kwh'] *= np.random.uniform(1.1, 1.3)

# Add random noise
df['energy_consumption_kwh'] += np.random.normal(0, 10, len(df))

# Save dataset
df.to_csv(r'C:\Users\gvija\Desktop\smart-city-energy-analytics\data', index=False)
df.head()
```
