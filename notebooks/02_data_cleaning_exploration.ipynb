```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/smart_city_energy.csv')

# Show first 5 rows
df.head()

# Check info
df.info()

# Check missing values
print(df.isnull().sum())

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Clean invalid data
df = df.dropna()
df = df[df['energy_consumption_kwh'] > 0]
df.reset_index(drop=True, inplace=True)

# Confirm cleaning
df.info()
```
